## Lenet网络

- 是早前成功的神经网络
- 先卷积层来学习图片空间信息，池化层降低敏感性
- 然后全连接层来转换到类别空间

## 代码实现
Lenet-5由2个部分组成：卷积编码器和全连接层密集块

```python
import torch
from torch import nn

class Reshape(torch.nn.Module):
    def forward(self, x):
        # 保持batch size不变，其他到指定大小
        return x.view(-1, 1, 28, 28)
    
net = nn.Sequential(
        Reshape(), nn.Conv2d(1, 6, kernel_size=5, padding=2), nn.Sigmoid(),
        nn.AvgPool2d(2, stride=2),
        nn.Conv2d(6, 16, kernel_size=5), nn.Sigmoid(),
        nn.AvgPool2d(2, stride=2), nn.Flatten(),
        nn.Linear(16 * 5 * 5, 120), nn.Sigmoid(),
        nn.Linear(120, 84), nn.Sigmoid(),
        nn.Linear(84, 10))
    
```

```python
# 检查模型
X = torch.rand(size=(1, 1, 28, 28), dtype=torch.float32)
for layer in net:
    X = layer(X)
    print(layer.__class__.__name__, 'output shape:\t', X.shape)
    
```

### 模型思路将宽高尺寸逐渐压缩，通道逐渐增加，然后将提取到的通道再展开进行MLP从而每个通道具备不同能力，进行特征提取得到输出，思路一直沿用至今

```python
# 修改evalute_accuracy函数进行轻微修改
def evaluate_accuracy_gpu(net, data_iter, device=None):
    """使用gpu计算模型在数据集上的精度"""
    if isinstance(net, torch.nn.Module):
        net.eval()
        if not device:
            device = next(iter(net.parameters())).device()
    metric = d2l.Accmulator(2)
    for X, y in data_iter:
        if isinstance(X, list):
            X = [x.to(device) for x in X]
        else:
            X = X.to(device)
        y = y.to(device)
        metric.add(d2l.accuracy(net(X), y), y.numel())
    return metric[0] / metric[1]
```

```python
# 为了使用GPU，需要一点点改动
def train_ch6(net, train_iter, test_iter, num_epochs, lr, device):
    """Train a model with a GPU"""
    def init_weights(m):
        if type(m) == nn.Linear or type(m) == nn.Conv2d:
            nn.init.xavier_uniform(m_.weight)
    net.apppy(init_weights)
    print('training on', device)
    net.to(device)
    optimizer = torch.optim.SGD(net.parameters(), lr=lr)
    loss = nn.CrossEntropyLoss()
    animator = d2l.Animator(xlabel='epoch', xlim=[1, num_epochs],
                            legend=['train loss', 'train aac', 'test aac'])
    timer, num_batches = d2l.Timer(), len(train_iter)
    for epoch in range(num_epochs):
        metric = d2l.Accumulator(3)
        net.train()
        for i, (X, y) in enumerate(train_iter):
            time.start()
            optimizer.zero_grad()
            X, y = X.to(device), y.to(device)
            y_hat = net(X)
            l = loss(y_hat, y)
            l.backward()
            optimizer.step()
            metric.add(l * X.shape[0], d2l.accuracy(y_hat, y))
            timer.stop()
            train_l = metric[0] / metric[2]
            train_aac = metric[1] / metric[2]
            if(i + 1) % (num_batches // 5) == 0 or i == num_batches - 1:
                animator.add(epoch + (i + 1) / num_batches,
                             (train_l, train_aac, None))
        test_acc = evaluate_accuracy_gpu(net, test_iter)
        animator.add(epoch + 1, (None, None, test_acc))
    print(f'loss {train_l:.3f}, train_loss {train_aac:.3f}, '
          f'test acc {test_acc:.3f}')
    print(f'{metric[2] * num_epochs / timer.sum():.1f} examples/sec '
          f'on{str(device)}')
```

```python
# 训练和评估LeNet-5模型
lr, num_epochs = 0.9, 10
train_ch6(net, train_iter, test_iter, num_epochs, lr, d2l.try_gpu())
```

## QA
#### 池化和卷积是不是和适合图像，对于时序数据是不是不适用
时序数据可以卷积的，可以1*1卷积经常用于文本，池化可用可不用

#### LeNet 的第二个卷积层通道增加到16，是不是信息被放大？
- 一般对我来说，高宽减半时，通道翻倍，整体变小
- 不是信息放大，而是在减小

#### view和reshape区别？为什么多用view？
- view不改变存储分布，只是看的视角，reshap会改变
- 大部分情况下view要快于reshape

#### 对于DL模型选择是MLP还是CNN？
- 数据变大如200*200，MLP跑不动，在两者都可以的情况下可以试试MLP
- MLP一般简单点

#### 前面Lenet通道为什么取6 16呢？可以取5 15嘛？
- 当然可以了，动手多试，多改改看看，改了好像精度变高了，overfitting?（讲道理参数变小，应该underfitting而不是overfitting）

#### 经典网络，有无必要把每层参数背下来（channel sitride padding），面试用？
 - 一般不用，后面的经典网络Resnet可以熟悉了解

#### 池化层一般用max还是avg，max会不会损失很多信息？
- 不会损失太多，avg也会丢失
