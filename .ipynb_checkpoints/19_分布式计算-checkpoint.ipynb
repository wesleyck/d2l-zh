# 分布式计算


- 尽量本地做通讯，偶尔跟内存做下通讯
- 少去不同机器间通讯
- 通讯速度逐次递减

### 计算一个小批量
假设2机器，每个机器双卡，batch 100
- 每个计算服务器读取小批量中的一块，单机器50，
- 然后单机器复制到GPUS上，25/卡
- 每个worker从参数服务器获取模型（单机器要一起）
- 复制参数到每个GPU上
- 每个GPU算自己梯度
- 将所有GPU上的梯度求和（先在主内存中，本地上做all_reduce加起来之后再发出去）
- 梯度传回服务器
- 每个服务器对梯度求和并更新参数

### 同步SGD
- 这里每个worker都是同步计算一个批量，称为同步SGD
- 假设有n个GPU，每个GPU每次处理b个样本，那么同步SGD = 单GPU运行nb的SGD
- 理想情况下，n个GPU可以得到相对单GPU的n倍加速

### 性能
- t1=在单个GPU上计算b个样本梯度的时间
- 假设有m个参数，一个worker每次发送和接受m个参数、梯度
    - t2 = 发送和接受所用时间
- 每个批量的计算时间max(t1,t2)
    - 选取足够大的b使得t1>t2(一般20-30%)
    - 增加b或n导致更大的批量大小，导致需要更多计算来得到给定的模型精度
- 所有的并行，单机多卡等等的核心性能瓶颈在收发数据

### 性能权衡


### 实践的建议
- 使用一个大数据集
- 需要好的GPU-GPU（主板）和机器-机器带宽（好的网线、交换机）
- 高效的数据读取和预处理（如随机选视频中的几个帧，比算一遍还慢，一定要看读取速度，有必要可以挪到GPU上算，如视频编解码）
- 模型需要有好的计算（FLOP）通讯（model size）比
    - Inception > Resnet > AlexNet
- 使用足够大的批量大小来得到更好的系统性能
- 使用高效的优化算法来对应大批量大小

#### 总结
- 分布式同步数据并行是多GPU数据并行在多机器上的拓展
- 网络通讯通常是瓶颈
- 需要注意使用特别大的批量大小时收敛效率
- 更复杂的分布式有异步、模型并行

## QA
#### 每个参数服务器求完·梯度之后，还要汇总到一个主服务器，然后再分发给各个参数服务器，然后再传给计算节点吗？
- 不是的，其实没有主服务器这么一说，如果参数服务器有三台，每个负责1/3，三个拼起来就算一个，每个worker向三个服务器要数据，然后本地汇总

#### 为什么batch size变大，训练的有效性曲线是下降的？
- 可以理论解释的，极端情况下，一个数据集样本全是同样的样本，那么size 1时候，需要100次迭代，完成精度，那么size*10时候，还是需要100次迭代(每次都是同样的梯度)，但却需要算1k次，相对来说性能下降
- 大部分情况下，数据不会完全一样，但是size越大，可能同一批数据中数据多样性越低，相对的样本梯度用处就不太大，就浪费了，差不多了图片梯度贡献差不多，如mnist基本128和512的数据多样性差不多，那肯定选少的

#### 分布式多个GPU增加batch size，是不是超过2048就不好了？
- 看你的数据和优化算法，如imagenet可以做到10w，好的优化算法也可以或者更大
- 一般看你类别，比如图片集n各类，最好你的batch size不要超过10*n
- 一般来说不用担心，没那么多机器，一般尽量加到GPU可以用的，然后有闲置机器，尽量加到一定情况就不用
